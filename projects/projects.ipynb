## Dataset
[EchoNet-Dynamic dataset](https://echonet.github.io/dynamic/index.html#dataset)

## Prerequisites
1. Download the dataset and unzip it to get a folder named `a4c-video-dir`. This is the dataset used for training and validation.
2. `git clone https://github.com/echonet/dynamic.git`
3. `cd dynamic`
4. Remove `sklearn==0.0` from `requirements.txt`, then run `pip install --user .`
5. Install sci-kit learn: `pip install scikit-learn`

## How to Use This Model
1. Place this `projects` folder under the `echonet` folder.
2. Add the following code to `echonet/utils/segmentation.py`:
```python
import echonet.projects.fusion_model as fusion_model
```
3. Around line 120 in segmentation.py, locate:
```python
    model = torchvision.models.segmentation.__dict__[model_name](pretrained=pretrained, aux_loss=True)
```
Replace it with:
```python
    model = fusion_model.FusionModel(pretrained=pretrained)
```
After replacing, you will be using the Fusion Model for image segmentation.
